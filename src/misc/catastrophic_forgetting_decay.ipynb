```python
import pandas as pd
import math
import matplotlib.pyplot as plt



MAX_MEM = 2000
decay_rate = -math.log(0.001) / MAX_MEM

pts = []
for mem in range(MAX_MEM):
    weight = 3 * math.exp(-decay_rate * mem)
    pts.append(weight)

plt.figure(figsize=(10, 6))
plt.plot(range(MAX_MEM), pts)
plt.xlabel('Time Step')
plt.ylabel('Weight')
plt.title('Weight Decay over Time Steps')
plt.grid(True)
plt.show()
```
