```python
import os,shutil
import pandas as pd
import re
```

## Remove noise

```python
labels = pd.read_csv('C:\\Users\\franc\\Desktop\\TESI\\SML_thesis_line_follower_robot\\webots\\data\\data\\sensors_data\\sensor_data 2024-08-20_19-18-23.csv',names=[str(i) for i in range(8)])['7']

sensors_data = pd.read_csv('C:\\Users\\franc\\Desktop\\TESI\\SML_thesis_line_follower_robot\\webots\\data\\data\\sensors_data\\sensor_data 2024-08-20_19-18-23.csv',names=[str(i) for i in range(8)]).drop('7',axis=1)
```

```python
row_means = sensors_data.mean(axis=1)
indices_with_zero_mean = sensors_data.index[row_means == 1000]
indices_with_zero_mean_list = indices_with_zero_mean.tolist()


labels = labels.drop(indices_with_zero_mean_list)
sensors_data = sensors_data.drop(indices_with_zero_mean_list)
```

```python
image_folder = 'C:\\Users\\franc\\Desktop\\TESI\\SML_thesis_line_follower_robot\\webots\\data\\images\\'

image_files = [f for f in os.listdir(image_folder) if os.path.isfile(os.path.join(image_folder, f))]

for image_file in image_files:
    if int(re.findall(r"\d+", image_file)[0]) in indices_with_zero_mean_list:
        os.remove(os.path.join(image_folder, image_file))
```

```python
labels.value_counts()
```

```python
import random
num_zeros_to_remove = int(labels.value_counts()[0]*0.85)
zero_indices = labels[labels == 0].index.tolist()
indices_to_remove = random.sample(zero_indices, num_zeros_to_remove)
```

```python
sensors_data = sensors_data.drop(indices_to_remove)
```

```python
labels = labels.drop(indices_to_remove)
```
