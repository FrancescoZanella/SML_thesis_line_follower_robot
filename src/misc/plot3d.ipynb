```python
import plotly.express as px
import pandas as pd

```

```python
df = pd.read_csv(r'C:\Users\franc\Desktop\TESI\SML_thesis_line_follower_robot\tmp\e-puck\data\data\sensors_data\sensors_data.csv')
```

### Plot data without drift


```python
df['color'] = df['drift'].map({1: 'red', 0: 'yellow'})
fig = px.scatter_3d(df, x='sensor0', y='sensor1', z='sensor2', 
                    color='color', opacity=0.7,color_discrete_map={'red': 'red', 'yellow': 'yellow'})


# Imposta le etichette degli assi e il titolo
fig.update_layout(scene=dict(xaxis_title='sensor0', yaxis_title='sensor1', zaxis_title='sensor2'),
                  title='Sensor Data with Drift Indication')

# Aggiorna la legenda
fig.update_layout(legend_title_text='Drift')
fig.for_each_trace(lambda t: t.update(name = 'Drift' if t.name == 'red' else 'No Drift'))

# Mostra il grafico
fig.show()
```
