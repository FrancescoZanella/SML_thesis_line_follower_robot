```python
import pandas as pd

df = pd.read_csv(r'C:\Users\franc\Desktop\TESI\SML_thesis_line_follower_robot\webots\data\data\sensors_data\umap_20_sensor_data_2024-08-23_18-01-07.csv')
df
```

# remove noise

```python
n_columns = ['sensor0','sensor1','sensor2','sensor3','sensor4','sensor5','sensor6']
row_means = df[n_columns].mean(axis=1)
indices_with_zero_mean = df[n_columns].index[row_means == 1000]
indices_with_zero_mean_list = indices_with_zero_mean.tolist()

df = df.drop(indices_with_zero_mean_list)

print(f'Removed {len(indices_with_zero_mean_list)} rows')
df
```

```python
df.to_csv(r'C:\Users\franc\Desktop\TESI\SML_thesis_line_follower_robot\webots\data\data\sensors_data\cleaned.csv',index=False)
```

## Sampling

```python
from imblearn.under_sampling import RandomUnderSampler

X, y = df.drop('target',axis=1),df['target']
undersample = RandomUnderSampler(sampling_strategy='majority')
X_over, y_over = undersample.fit_resample(X, y)
```

```python
df_sampled = pd.DataFrame(X_over)
df_sampled['target'] = y_over
df_sampled = df_sampled.sort_index()
df_sampled
```

```python
df_sampled.to_csv(r'C:\Users\franc\Desktop\TESI\SML_thesis_line_follower_robot\webots\data\data\sensors_data\sampled_umap_20_sensor_data_2024-08-23_18-01-07.csv',index=False)
```
