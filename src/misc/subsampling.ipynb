```python
import pandas as pd

df = pd.read_csv(r'C:\Users\franc\Desktop\TESI\SML_thesis_line_follower_robot\webots\data\data\sensors_data\sensor_data 2024-08-23_11-43-13.csv',names=['sensor0','sensor1','sensor2','sensor3','sensor4','sensor5','sensor6','target'])
df
```

```python
from imblearn.under_sampling import RandomUnderSampler

X, y = df.drop('target',axis=1),df['target']
undersample = RandomUnderSampler(sampling_strategy='majority')
X_over, y_over = undersample.fit_resample(X, y)
```

```python
df_sampled = pd.DataFrame(X_over)
df_sampled['target'] = y_over
df_sampled = df_sampled.sort_index()
df_sampled
```

```python
df_sampled.to_csv(r'C:\Users\franc\Desktop\TESI\SML_thesis_line_follower_robot\webots\data\data\sensors_data\sampled.csv',index=False,header=False)
```
