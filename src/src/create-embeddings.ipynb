```python
from tensorflow.keras import layers as tfkl
from tensorflow import keras as tfk
import tensorflow as tf
from PIL import Image
import os
import numpy as np
```

## Load model

```python
model = tfk.models.load_model('/kaggle/input/my_net/keras/default/1/my_net.keras')
```

## Create embeddings

```python
image_folder = '/kaggle/input/images/images'
embeddings = []
image_paths = []
```

```python
for img_name in os.listdir(image_folder):
    
    img_path = os.path.join(image_folder, img_name)
    
    image = Image.open(img_path)
    
    image = tf.cast(image, tf.float32) / 255.0
    
    flatten_layer = tf.keras.Sequential(model.layers[:8])
    
    embedding = flatten_layer(tf.expand_dims(image, axis=0))
        
    embedding = tf.squeeze(embedding, axis=0)
    
    embeddings.append(embedding.numpy())
    image_paths.append(img_path)
    
    
embeddings = np.array(embeddings)
np.save('image_embeddings.npy', embeddings)
```
